# Pre-trained transormer models

```python
import sys
import subprocess
import transformers
from transformers import pipeline
```

```python
sent_pipeline=pipeline("sentiment-analysis")
```

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

```python
print(sentence_1)
sent_pipeline(sentence_1)
```

```python
te1=sent_pipeline(sentence_1)
te2=sent_pipeline(sentence_2)
te3=sent_pipeline(sentence_3)
te4=sent_pipeline(sentence_4)


```

```python
print([sub['label'] for sub in te1])
print([sub['label'] for sub in te2])
print([sub['label'] for sub in te3])
print([sub['label'] for sub in te4])
```
