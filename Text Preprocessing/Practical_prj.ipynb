#  Practical

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
import matplotlib.pyplot as pt
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')
```

### Load data

```python
data = pd.read_csv("/content/tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

### Lowercase

```python
data['review_lower'] = data['Review'].str.lower()
```

```python
data.head()
```

### Stop word removal

```python
en_stopwords = stopwords.words('english')
en_stopwords.remove("not")
```

```python
data['review_nostopwords'] = data['review_lower'] \
                                .apply(lambda x: ' ' \
                                .join([word for word \
                                in x.split() if word \
                                not in (en_stopwords)]))
```

```python
data.head()
```

```python
data['review_nostopwords'][0]
```

### Punctuation

We want to remove punctation but we notice we have a few reviews with the symbol * instead of the word star. We want to keep this as it might add meaning to the review. We can do this using what we learned about regular expressions.

```python
data['review_nostop_nopunct'] = data['review_nostopwords'].apply(lambda x:
                                    re.sub(r"[*]", "star",
                                    x))
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'] = data['review_nostop_nopunct'].apply(lambda x:
                                    re.sub(r"([^\w\s])",  "",
                                    x))
```

```python
data.head()
```

```python
data['review_no_stopwords_no_punct'][0]
```

### Tokenizing

```python
data['tokenized'] = data['review_no_stopwords_no_punct'].apply(lambda x:
                                                               word_tokenize(x))
```

```python
data.head()
```

```python
data['tokenized'][0]
```

### Stemming

```python
ps = PorterStemmer()
```

```python
data["stemmed"] = data["tokenized"].apply(lambda x:
                  [ps.stem(i)  for i in x])
```

```python
data.head()
```

```python
data['stemmed'][0]
```

### Lemmatization

```python
lemmatizer = WordNetLemmatizer()
```

```python
data["lemmatized"] = data["tokenized"].apply(lambda x: \
                    [lemmatizer.lemmatize(i) \
                     for i in x])
```

```python
data['lemmatized'][0]
```

```python
data.head()
```

## N-grams

```python
tokens_clean = sum(data['lemmatized'], [])
```

```python
# unigrams: n=1
unigrams = (pd.Series(nltk.ngrams(tokens_clean, 1))
            .value_counts())

print(unigrams)
```

```python
# bigrams: n=2
bigrams = (pd.Series(nltk.ngrams(tokens_clean, 2))
           .value_counts())
print(bigrams)
```

```python
ngrams_4 = (pd.Series(nltk.ngrams(tokens_clean, 4))
            .value_counts())
print(ngrams_4)
```
